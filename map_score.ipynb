```python
import pandas as pd

# Load the dataset
df = pd.read_csv('all_rt_scores.csv')

# Define a function to map scores to categories
def map_score_to_category(score):
    if pd.isnull(score):
        return None  # Return None or a specific string like 'unknown' if the score is NaN
    elif score < 60:
        return 'rotten'
    elif score < 75:
        return 'fresh'
    else:
        return 'certified fresh'

# Apply the function to the 'CriticScore' column
df['CriticScoreCategory'] = df['CriticScore'].apply(map_score_to_category)
df = df.dropna(subset=['CriticScore'])

# Save the modified DataFrame to a new CSV file
df.to_csv('modified_all_rt_scores.csv', index=False)

df.head()  # Display the first few rows of the modified DataFrame
```
